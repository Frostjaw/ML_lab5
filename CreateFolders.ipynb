```python
import os, shutil, glob

# The path to the directory where the original
# dataset was uncompressed
original_dataset_train_o_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET_original/TRAIN/O'
original_dataset_train_r_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET_original/TRAIN/R'
original_dataset_test_o_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET_original/TEST/O'
original_dataset_test_r_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET_original/TEST/R'

# Directory with our training organic pictures
train_o_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/train/O'

# Directory with our training recyclable pictures
train_r_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/train/R'

# Directory with our validation organic pictures
validation_o_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/validation/O'

# Directory with our validation recyclable pictures
validation_r_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/validation/R'

# Directory with our test organic pictures
test_o_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/test/O'

# Directory with our test recyclable pictures
test_r_dir = 'D:/Учеба/4 course/машинка/Лабы/lab5_Мошняков/DATASET/test/R'

# Copy first 1000 organic images to train_o_dir
fnames = glob.glob(os.path.join(original_dataset_train_o_dir,"*.jpg"))[1:1000]
for fname in fnames:
    src = os.path.join(original_dataset_train_o_dir, os.path.basename(fname))
    dst = os.path.join(train_o_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)

# Copy next 500 organic images to validation_o_dir
fnames = glob.glob(os.path.join(original_dataset_train_o_dir,"*.jpg"))[1000:1500]
for fname in fnames:
    src = os.path.join(original_dataset_train_o_dir, os.path.basename(fname))
    dst = os.path.join(validation_o_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)
    
# Copy next 500 organic images to test_o_dir
fnames = glob.glob(os.path.join(original_dataset_test_o_dir,"*.jpg"))[1:500]
for fname in fnames:
    src = os.path.join(original_dataset_test_o_dir, os.path.basename(fname))
    dst = os.path.join(test_o_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)
    
# Copy first 1000 recyclable images to train_r_dir
fnames = glob.glob(os.path.join(original_dataset_train_r_dir,"*.jpg"))[1:1000]
for fname in fnames:
    src = os.path.join(original_dataset_train_r_dir, os.path.basename(fname))
    dst = os.path.join(train_r_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)
    
# Copy next 500 recyclable images to validation_r_dir
fnames = glob.glob(os.path.join(original_dataset_train_r_dir,"*.jpg"))[1000:1500]
for fname in fnames:
    src = os.path.join(original_dataset_train_r_dir, os.path.basename(fname))
    dst = os.path.join(validation_r_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)
    
# Copy next 500 recyclable images to test_r_dir
fnames = glob.glob(os.path.join(original_dataset_test_r_dir,"*.jpg"))[1:500]
for fname in fnames:
    src = os.path.join(original_dataset_test_r_dir, os.path.basename(fname))
    dst = os.path.join(test_r_dir, os.path.basename(fname))
    shutil.copyfile(src, dst)
```
